```python
temps = [28, 32, 31, 29, 35, 27, 30]
max_temp = max(temps)
min_temp = min(temps)
avg_temp = sum(temps) / len(temps)
above_30 = [i for i in range(len(temps)) if temps[i] > 30]
days_above_30 = len(above_30)
exact_temps = [temps[day] for day in above_30]
exact_days = [day + 1 for day in above_30]
```
